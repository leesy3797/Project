```python
df = pd.read_csv('/content/drive/MyDrive/ASAC3기_이상탐지/Data/Machine_Signal_30_train.csv')
df_y = pd.read_csv('/content/drive/MyDrive/ASAC3기_이상탐지/Data/Machine_Signal_Sample3_Y.csv')
```

```python
df.shape, df_y.shape
```

```python
valid_machine_list = list(df_y.machine_id.unique())
valid_machine_list
```

```python
df_3=df[df.machine_id.apply(lambda x : x in valid_machine_list)]
```

```python
df_merge = pd.merge(df_3, df_y[['ckey', 'Y']], on='ckey', how='left')
```

```python
df_merge['Y'] = df_merge['Y'].fillna(0)
```

```python
df_merge.to_csv('/content/drive/MyDrive/ASAC3기_이상탐지/Code/5주차/이승용/train.csv', index=False)
```
